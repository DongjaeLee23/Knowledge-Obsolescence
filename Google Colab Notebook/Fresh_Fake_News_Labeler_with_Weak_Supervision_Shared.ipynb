## Training and Testing

```python
import torch

device = "cuda" if torch.cuda.is_available() else "cpu"
```

```python
import sys
import json
import os.path as osp
sys.path.append('/content/KO')
from config import interpret_config
from datasets.covid19_tweets import Covid19Tweets
from torch.utils.data import DataLoader
from model import build_model
from trainer import Trainer


def train(config, train_datapath=''):
    if train_datapath != '':
        config["data_param"]["train_datapath"] = train_datapath
    config = interpret_config(config)
    # NOTE: We can use torch.utils.data.ConcatDataset if we want to combine multiple datasets
    # https://saturncloud.io/blog/pytorch-concatenating-datasets-before-using-dataloader/
    dataset = Covid19Tweets(osp.join(config['data_param']['data_root'], f'time_sorted/{config["data_param"]["train_datapath"]}.csv'), tokenizer=config['tokenizer'])
    dataloader = DataLoader(dataset, batch_size=config['data_param']['batch_size'], shuffle=True, num_workers=2)
    model = build_model(config)
    trainer = Trainer(config, model, dataloader, None, None, f'{config["model"]}_{config["data_param"]["train_datapath"]}')
    trainer.train()

def test(config, train_datapath='', test_datapath='', checkpoint_path=''):
    if train_datapath != '':
        config["data_param"]["train_datapath"] = train_datapath
    if test_datapath != '':
        config["data_param"]["test_datapath"] = test_datapath
    if checkpoint_path == '':
        checkpoint_path = f'/content/drive/My Drive/KO/checkpoints/{config["model"]}_{config["data_param"]["train_datapath"]}.pt'
    config = interpret_config(config)
    config["trainer_param"]["epochs"] = 0
    dataset = Covid19Tweets(osp.join(config['data_param']['data_root'], f'time_sorted/{config["data_param"]["test_datapath"]}.csv'), tokenizer=config['tokenizer'])
    dataloader = DataLoader(dataset, batch_size=config['data_param']['batch_size'], shuffle=False, num_workers=2)
    model = build_model(config)
    trainer = Trainer(config, model, None, None, dataloader, f'{config["model"]}_{config["data_param"]["train_datapath"]}_{config["data_param"]["test_datapath"]}')
    checkpoint = torch.load(checkpoint_path, map_location=device)
    trainer.module.model.load_state_dict(checkpoint['model_state_dict'])
    trainer.test()
```

# Models

```python
"""Adapted from https://github.com/huggingface/transformers/blob/v4.34.0/src/transformers/models/albert/modeling_albert.py#L1025"""
import torch.nn as nn
from transformers import AlbertModel, AlbertPreTrainedModel

class Albert(AlbertPreTrainedModel):
    def __init__(self, config):
        super().__init__(config)
        self.config = config
        self.encoder = AlbertModel.from_pretrained("albert-base-v2", config=self.config)
        self.dropout = nn.Dropout(self.config.classifier_dropout_prob)
        self.classifier = nn.Linear(self.config.hidden_size, self.config.num_labels)

        # Initialize weights and apply final processing
        self.post_init()

    def forward(self, x, attention_mask=None, token_type_ids=None, position_ids=None, head_mask=None):
        outputs = self.encoder(x,
                            attention_mask=attention_mask,
                            token_type_ids=token_type_ids,
                            position_ids=position_ids,
                            head_mask=head_mask)

        pooled_output = outputs[1]
        pooled_output = self.dropout(pooled_output)
        logits = self.classifier(pooled_output)
        return logits
```

```python
"""Adapted from https://github.com/huggingface/transformers/blob/v4.34.0/src/transformers/models/bert/modeling_bert.py#L1519"""
import torch.nn as nn
from transformers import BertModel, BertPreTrainedModel

class Bert(BertPreTrainedModel):
    def __init__(self, config):
        super().__init__(config)
        self.config = config
        self.encoder = BertModel.from_pretrained("bert-base-uncased", config=self.config)
        classifier_dropout = (
            self.config.classifier_dropout if self.config.classifier_dropout is not None else self.config.hidden_dropout_prob
        )
        self.dropout = nn.Dropout(classifier_dropout)
        self.classifier = nn.Linear(self.config.hidden_size, self.config.num_labels)

        # Initialize weights and apply final processing
        self.post_init()

    def forward(self, x, attention_mask=None, token_type_ids=None, position_ids=None, head_mask=None):
        outputs = self.encoder(x,
                            attention_mask=attention_mask,
                            token_type_ids=token_type_ids,
                            position_ids=position_ids,
                            head_mask=head_mask)

        pooled_output = outputs[1]
        pooled_output = self.dropout(pooled_output)
        logits = self.classifier(pooled_output)
        return logits
```

```python
import torch.nn as nn
from transformers import GPT2Model, GPT2PreTrainedModel

class GPT2(GPT2PreTrainedModel):
    def __init__(self, config):
        super().__init__(config)
        self.config = config
        self.encoder = GPT2Model.from_pretrained("gpt2", config=self.config)
        self.dropout = nn.Dropout(self.config.attention_probs_dropout_prob)
        self.classifier = nn.Linear(self.config.hidden_size, self.config.num_labels)

        # Initialize weights and apply final processing
        self.post_init()

    def forward(self, x, attention_mask=None, token_type_ids=None, position_ids=None, head_mask=None):
        outputs = self.encoder(x,
                            attention_mask=attention_mask,
                            token_type_ids=token_type_ids,
                            position_ids=position_ids,
                            head_mask=head_mask)

        pooled_output = outputs[0][:, 0, :]  # Take the representation of the [CLS] token
        pooled_output = self.dropout(pooled_output)
        logits = self.classifier(pooled_output)
        return logits
```

# Dataloader

```python
# From https://github.com/asuprem/EdnaML/blob/a389c2b9d2fa9f21422c4ae5f961153035e8bfd2/src/ednaml/generators/HFGenerator.py#L498
def convert_to_features(text, labels, maxlen, tokenizer):
    """Preprocess the dataset into text features.

    Args:
        text: list of text data
        labels: list of labels corresponding to text
        maxlen (_type_): _description_
    """
    features = []
    for idx, sample in enumerate(zip(text, labels)):
        tokens = tokenizer.tokenize(sample[0])
        if len(tokens) > maxlen - 2:
            tokens = tokens[0:(maxlen - 2)]
        finaltokens = ["[CLS]"]
        token_type_ids = [0]
        for token in tokens:
            finaltokens.append(token)
            token_type_ids.append(0)
        finaltokens.append("[SEP]")
        token_type_ids.append(0)

        input_ids = tokenizer.convert_tokens_to_ids(finaltokens)
        attention_mask = [1]*len(input_ids)
        input_len = len(input_ids)
        while len(input_ids) < maxlen:
            input_ids.append(0)
            attention_mask.append(0)
            token_type_ids.append(0)

        assert len(input_ids) == maxlen
        assert len(attention_mask) == maxlen
        assert len(token_type_ids) == maxlen

        features.append(
            (torch.tensor(input_ids), torch.tensor(attention_mask), torch.tensor(token_type_ids), input_len, *sample[1:])
        )

    return features
```

You can use the following two labeling functions as reference while creating your labeling function. The Snorkel pipeline requires at least three labeling functions. In our methodology, we have at least four datasets to train expert models. Usually Bert models perform better than Albert ones, but the former might require more computational power. Feel free to choose any models you like in the experiments. You can also try other customized models.

```python
@labeling_function()
def lf_expert_albert_1_2020(x):
    """Follow this format to add other expert models"""
    model_config = copy.deepcopy(albert_config["model_param"])
    num_labels = albert_config["data_param"]["num_classes"]
    model_config["num_labels"] = num_labels
    model_config = AlbertConfig(**model_config)
    model = Albert(model_config).to(device)
    checkpoint = torch.load(albert_1_2020_checkpoint, map_location=device)
    model.load_state_dict(checkpoint['model_state_dict'])
    tokenizer = AlbertTokenizer.from_pretrained('albert-base-v2')
    input = convert_to_features([x.text], [0], maxlen=tokenizer.model_max_length, tokenizer=tokenizer)
    all_input_ids, all_attention_mask, all_token_type_ids, all_input_len, all_labels = input[0]
    output = model(all_input_ids.to(device).reshape(1, -1), token_type_ids=all_token_type_ids.to(device).reshape(1, -1), attention_mask=all_attention_mask.to(device).reshape(1, -1))
    preds = torch.argmax(output.cpu(), dim=1)
    return int(preds)
```
